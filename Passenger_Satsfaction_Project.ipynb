```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np

# กำหนดช่วงของค่า K ที่ต้องการทดสอบ
k_range = range(2, 11)

# เก็บค่า SSE, Silhouette Score และ Davies-Bouldin Index สำหรับแต่ละค่า K
sse_values = []
silhouette_scores = []
db_scores = []

# วนลูปเพื่อคำนวณค่าต่างๆ สำหรับแต่ละค่า K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # สร้าง KMeans model
    kmeans.fit(df_cluster)  # Fit model กับข้อมูล

    # เก็บค่า SSE
    sse_values.append(kmeans.inertia_)

    # คำนวณและเก็บค่า Silhouette Score
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(df_cluster, labels))

    # คำนวณและเก็บค่า Davies-Bouldin Index
    db_scores.append(davies_bouldin_score(df_cluster, labels))

# แสดงค่าต่างๆ สำหรับแต่ละค่า K
for i, k in enumerate(k_range):
    print(f"K={k}, SSE={sse_values[i]:.2f}, Silhouette Score={silhouette_scores[i]:.2f}, Davies-Bouldin Index={db_scores[i]:.2f}")
```

##K-Means with Silhouette Method

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# กำหนดช่วงของค่า K ที่ต้องการทดสอบ
k_range = range(2, 11)  # ทดสอบ K ตั้งแต่ 2 ถึง 10

# เก็บค่า Silhouette Score สำหรับแต่ละค่า K
silhouette_scores = []

# วนลูปเพื่อคำนวณ Silhouette Score สำหรับแต่ละค่า K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # สร้าง KMeans model
    cluster_labels = kmeans.fit_predict(df_cluster)  # ทำการ clustering
    silhouette_avg = silhouette_score(df_cluster, cluster_labels)  # คำนวณ Silhouette Score
    silhouette_scores.append(silhouette_avg)  # เก็บค่า Silhouette Score

# พล็อตกราฟ Silhouette Score
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')

# หาค่า K ที่ให้ Silhouette Score สูงสุด
optimal_k = k_range[silhouette_scores.index(max(silhouette_scores))]
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Optimal K ({optimal_k})')  # เส้นตัดสินใจ

plt.legend()
plt.show()

print(f"Optimal number of clusters: {optimal_k}")
```

##Davie bouldin

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt

# กำหนดช่วงของค่า K ที่ต้องการทดสอบ
k_range = range(2, 11)  # ทดสอบ K ตั้งแต่ 2 ถึง 10

# เก็บค่า Davies-Bouldin Index สำหรับแต่ละค่า K
db_scores = []

# วนลูปเพื่อคำนวณ Davies-Bouldin Index สำหรับแต่ละค่า K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # สร้าง KMeans model
    cluster_labels = kmeans.fit_predict(df_cluster)  # ทำการ clustering
    db_score = davies_bouldin_score(df_cluster, cluster_labels)  # คำนวณ Davies-Bouldin Index
    db_scores.append(db_score)  # เก็บค่า Davies-Bouldin Index

# พล็อตกราฟ Davies-Bouldin Index
plt.plot(k_range, db_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Index')

# หาค่า K ที่ให้ Davies-Bouldin Index ต่ำสุด
optimal_k = k_range[db_scores.index(min(db_scores))]
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Optimal K ({optimal_k})')  # เส้นตัดสินใจ

plt.legend()
plt.show()

print(f"Optimal number of clusters: {optimal_k}")
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
# คำนวณ Silhouette Score
silhouette_avg = silhouette_score(df_cluster.drop('Cluster', axis=1), df_cluster['Cluster'])  # ไม่รวม column 'Cluster' ในการคำนวณ
print(f"Silhouette Score: {silhouette_avg:.3f}")

# คำนวณ Davies-Bouldin Index
davies_bouldin_avg = davies_bouldin_score(df_cluster.drop('Cluster', axis=1), df_cluster['Cluster']) # ไม่รวม column 'Cluster' ในการคำนวณ
print(f"Davies-Bouldin Index: {davies_bouldin_avg:.3f}")
```

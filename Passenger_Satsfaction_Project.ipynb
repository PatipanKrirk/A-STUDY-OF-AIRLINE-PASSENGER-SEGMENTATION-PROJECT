##Normalization

```python
from sklearn.preprocessing import StandardScaler

# à¸ªà¸£à¹‰à¸²à¸‡ StandardScaler
s_scaler = StandardScaler()

# à¸›à¸£à¸±à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰ Standardization
df_cluster = pd.DataFrame(s_scaler.fit_transform(df_cluster), columns=df_cluster.columns)
df_cluster.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming df_cluster is already loaded and contains the selected features

sse = []  # List to store SSE values

# Try different values of k (number of clusters)
k_values = range(1, 11)  # Example: Try k from 1 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)  # Create KMeans object
    kmeans.fit(df_cluster)  # Fit to your data
    sse.append(kmeans.inertia_)  # Append SSE to the list

# Plot the SSE values against the number of clusters (k)
plt.plot(k_values, sse, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# à¸„à¸³à¸™à¸§à¸“ SSE à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸ˆà¸³à¸™à¸§à¸™à¸„à¸¥à¸±à¸ªà¹€à¸•à¸­à¸£à¹Œ
sse = []
k_range = range(2, 11)  # à¸—à¸”à¸¥à¸­à¸‡à¸ˆà¸³à¸™à¸§à¸™à¸„à¸¥à¸±à¸ªà¹€à¸•à¸­à¸£à¹Œà¸•à¸±à¹‰à¸‡à¹à¸•à¹ˆ 2 à¸–à¸¶à¸‡ 10

# Fit KMeans à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸° K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_cluster)
    sse.append(kmeans.inertia_)  # à¸„à¹ˆà¸² inertia_ à¸„à¸·à¸­ SSE

# à¸žà¸¥à¹‡à¸­à¸•à¸à¸£à¸²à¸Ÿ Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o', linestyle='-', color='b')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.xticks(k_range)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np

# à¸à¸³à¸«à¸™à¸”à¸Šà¹ˆà¸§à¸‡à¸‚à¸­à¸‡à¸„à¹ˆà¸² K à¸—à¸µà¹ˆà¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸—à¸”à¸ªà¸­à¸š
k_range = range(2, 11)

# à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² SSE, Silhouette Score à¹à¸¥à¸° Davies-Bouldin Index à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
sse_values = []
silhouette_scores = []
db_scores = []

# à¸§à¸™à¸¥à¸¹à¸›à¹€à¸žà¸·à¹ˆà¸­à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¸•à¹ˆà¸²à¸‡à¹† à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # à¸ªà¸£à¹‰à¸²à¸‡ KMeans model
    kmeans.fit(df_cluster)  # Fit model à¸à¸±à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥

    # à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² SSE
    sse_values.append(kmeans.inertia_)

    # à¸„à¸³à¸™à¸§à¸“à¹à¸¥à¸°à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² Silhouette Score
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(df_cluster, labels))

    # à¸„à¸³à¸™à¸§à¸“à¹à¸¥à¸°à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² Davies-Bouldin Index
    db_scores.append(davies_bouldin_score(df_cluster, labels))

# à¹à¸ªà¸”à¸‡à¸„à¹ˆà¸²à¸•à¹ˆà¸²à¸‡à¹† à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
for i, k in enumerate(k_range):
    print(f"K={k}, SSE={sse_values[i]:.2f}, Silhouette Score={silhouette_scores[i]:.2f}, Davies-Bouldin Index={db_scores[i]:.2f}")
```

##K-Means with Silhouette Method

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# à¸à¸³à¸«à¸™à¸”à¸Šà¹ˆà¸§à¸‡à¸‚à¸­à¸‡à¸„à¹ˆà¸² K à¸—à¸µà¹ˆà¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸—à¸”à¸ªà¸­à¸š
k_range = range(2, 11)  # à¸—à¸”à¸ªà¸­à¸š K à¸•à¸±à¹‰à¸‡à¹à¸•à¹ˆ 2 à¸–à¸¶à¸‡ 10

# à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² Silhouette Score à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
silhouette_scores = []

# à¸§à¸™à¸¥à¸¹à¸›à¹€à¸žà¸·à¹ˆà¸­à¸„à¸³à¸™à¸§à¸“ Silhouette Score à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # à¸ªà¸£à¹‰à¸²à¸‡ KMeans model
    cluster_labels = kmeans.fit_predict(df_cluster)  # à¸—à¸³à¸à¸²à¸£ clustering
    silhouette_avg = silhouette_score(df_cluster, cluster_labels)  # à¸„à¸³à¸™à¸§à¸“ Silhouette Score
    silhouette_scores.append(silhouette_avg)  # à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² Silhouette Score

# à¸žà¸¥à¹‡à¸­à¸•à¸à¸£à¸²à¸Ÿ Silhouette Score
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')

# à¸«à¸²à¸„à¹ˆà¸² K à¸—à¸µà¹ˆà¹ƒà¸«à¹‰ Silhouette Score à¸ªà¸¹à¸‡à¸ªà¸¸à¸”
optimal_k = k_range[silhouette_scores.index(max(silhouette_scores))]
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Optimal K ({optimal_k})')  # à¹€à¸ªà¹‰à¸™à¸•à¸±à¸”à¸ªà¸´à¸™à¹ƒà¸ˆ

plt.legend()
plt.show()

print(f"Optimal number of clusters: {optimal_k}")
```

##Davie bouldin

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt

# à¸à¸³à¸«à¸™à¸”à¸Šà¹ˆà¸§à¸‡à¸‚à¸­à¸‡à¸„à¹ˆà¸² K à¸—à¸µà¹ˆà¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸—à¸”à¸ªà¸­à¸š
k_range = range(2, 11)  # à¸—à¸”à¸ªà¸­à¸š K à¸•à¸±à¹‰à¸‡à¹à¸•à¹ˆ 2 à¸–à¸¶à¸‡ 10

# à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² Davies-Bouldin Index à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
db_scores = []

# à¸§à¸™à¸¥à¸¹à¸›à¹€à¸žà¸·à¹ˆà¸­à¸„à¸³à¸™à¸§à¸“ Davies-Bouldin Index à¸ªà¸³à¸«à¸£à¸±à¸šà¹à¸•à¹ˆà¸¥à¸°à¸„à¹ˆà¸² K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # à¸ªà¸£à¹‰à¸²à¸‡ KMeans model
    cluster_labels = kmeans.fit_predict(df_cluster)  # à¸—à¸³à¸à¸²à¸£ clustering
    db_score = davies_bouldin_score(df_cluster, cluster_labels)  # à¸„à¸³à¸™à¸§à¸“ Davies-Bouldin Index
    db_scores.append(db_score)  # à¹€à¸à¹‡à¸šà¸„à¹ˆà¸² Davies-Bouldin Index

# à¸žà¸¥à¹‡à¸­à¸•à¸à¸£à¸²à¸Ÿ Davies-Bouldin Index
plt.plot(k_range, db_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Index')

# à¸«à¸²à¸„à¹ˆà¸² K à¸—à¸µà¹ˆà¹ƒà¸«à¹‰ Davies-Bouldin Index à¸•à¹ˆà¸³à¸ªà¸¸à¸”
optimal_k = k_range[db_scores.index(min(db_scores))]
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Optimal K ({optimal_k})')  # à¹€à¸ªà¹‰à¸™à¸•à¸±à¸”à¸ªà¸´à¸™à¹ƒà¸ˆ

plt.legend()
plt.show()

print(f"Optimal number of clusters: {optimal_k}")
```
